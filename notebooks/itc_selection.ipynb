```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# Load the file
df = pd.read_csv('../results/v6_final_consensus_hits_docking_merged.csv')

# Look at top 20 by binding energy
df_top = df.sort_values(by='Binding Energy (kcal/mol)', ascending=True).head(20).copy()

# Calculate MW and LogP to assess aqueous solubility (critical for ITC)
def get_properties(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        return Descriptors.MolWt(mol), Descriptors.MolLogP(mol)
    return None, None

df_top[['MW', 'LogP']] = df_top.apply(lambda row: pd.Series(get_properties(row['smiles'])), axis=1)

print(df_top[['pert_iname', 'Binding Energy (kcal/mol)', 'activity', 'MW', 'LogP']].to_string(index=False))
```
