```python
import pandas as pd

# Load the consensus file
df = pd.read_csv('final_hit_list.csv')

# Apply the filters
gold_hits = df[(df['activity'] > 0.9) & (df['Binding Energy (kcal/mol)'] < -9.0)]

# Sort by best binding energy
gold_hits = gold_hits.sort_values(by='Binding Energy (kcal/mol)')

# Display the hits
print(f"Found {len(gold_hits)} compounds meeting the criteria.")
print(gold_hits[['pert_iname', 'activity', 'Binding Energy (kcal/mol)']])

# Save specifically these leads
gold_hits.to_csv('top_leads_filtered.csv', index=False)
```
